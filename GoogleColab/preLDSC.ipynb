```python
def formatSumStatsForLDSC_PD (csvPath):
    df = pd.read_csv(csvPath)
    newDf = df[['SNP', 'A1', 'A2', 'b','se', 'N_cases', 'N_controls', 'p']].copy()
    newDf.insert(3, 'Zscore', newDf['b'] / newDf['se'], True)
    newDf.insert(4, 'N', newDf['N_cases'] + newDf['N_controls'], True)
    newDf = newDf.drop(['b','se'], axis = 1)
    newDf = newDf.rename(columns={ 'SNP' : 'snpid', 'p' : 'P-value'}) # Now have: snpid A1 A2 Zscore N P-value
    newDf.to_csv("/content/drive/MyDrive/Genetics_Research/nalls_datasetFormatted.csv", index=False)
    
def formatSumStatsForLDSC_MB (csvPath):
    df = pd.read_csv(csvPath)
    newDf = df[['rsID', 'eff.allele', 'ref.allele', 'beta','SE', 'N', 'pDerived']].copy()
    newDf.insert(3, 'Zscore', newDf['beta'] / newDf['SE'], True)
    newDf = newDf.drop(['beta','SE'], axis = 1)
    newDf = newDf.rename(columns={'rsID' : 'snpid', 'pDerived' : 'P-value','eff.allele' : 'A1', 'ref.allele' : 'A2'}) # Now have: snpid A1 A2 Zscore N P-value
    newDf.to_csv("/content/drive/MyDrive/Genetics_Research/bacEx_datasetFormatted.csv", index=False)
```

```python
# Arrays
snpIDsToReplaceInNallsSumStats = dfnalls['snpid'].to_list()
allsnpIDsMappedInList = list(allids.items())
```

```python
#not deleting if chr_pos in allsnpIDsMappedInList[pair][0] cant be converted to int bx X_12782 or something
from tqdm.notebook import tqdm

for pair in tqdm(range(len(allsnpIDsMappedInList))):
  allsnpIDsMappedInList[pair] = list(allsnpIDsMappedInList[pair])
  try:
    allsnpIDsMappedInList[pair][0] = int(allsnpIDsMappedInList[pair][0].replace("_",""))
  except ValueError or IndexError: # bc sometimes no "_" in allsnpIDsMappedInList[pair][0] or cant be converted to int
    pass
```

```python
def binary_search(arr, low, high, x):
 
    # Check base case
    if high >= low:
 
        mid = (high + low) // 2
 
        if isinstance(arr[mid][0], int):
          # If element is present at the middle itself
          if arr[mid][0] == x:
              return mid
  
          # If element is smaller than mid, then it can only
          # be present in left subarray
          elif arr[mid][0] > x:
              return binary_search(arr, low, mid - 1, x)
  
          # Else the element can only be present in right subarray
          else:
              return binary_search(arr, mid + 1, high, x)
  
    else:
        # Element is not present in the array
        return -1
```

```python
#More efficient way
from tqdm.notebook import tqdm
import time

start = time.time()
for i in tqdm(range(len(snpIDsToReplaceInNallsSumStats))):
  if not snpIDsToReplaceInNallsSumStats[i].startswith("rs") and not snpIDsToReplaceInNallsSumStats[i].startswith("."):
    chr_pos = snpIDsToReplaceInNallsSumStats[i]
    toFind = int(snpIDsToReplaceInNallsSumStats[i].replace("_","")) #converting chr_pos to int from snpIDsToReplaceInNallsSumStats
    #print(toFind)
    result = binary_search(allsnpIDsMappedInList, 0, len(allsnpIDsMappedInList) - 1, toFind)
    #print("index in allsnpIDsMappedInList:", result)
    
    if result != -1: #means there was a match, the result is index of allsnpIDsMappedInList where it was found, , -1 means no match
      try:
        snpIDsToReplaceInNallsSumStats[i] = allsnpIDsMappedInList[result][1] #sometimes allsnpIDsMappedInList[result][1] will be "."
        #print(chr_pos, "replaced to", allsnpIDsMappedInList[result][1])
      except TypeError: # when no match, result will equal None, so just bypass the current snp in Nalls data
        pass
end = time.time()
print("Time taken:", (end - start)/60, "mins")
```
