```python
lstOfSumResultsStartIndex = []

with open("/Volumes/T7Touch/NIHSummer2021/Data/LDSR_analysis/ldsc/ldsc_results.txt") as f:
   for i, line in enumerate(f):
        
       if "Summary of Genetic Correlation Results" in line:
            lstOfSumResultsStartIndex.append(i)
```

```python
len(lstOfSumResultsStartIndex)
```

```python
file = open("/Volumes/T7Touch/NIHSummer2021/Data/LDSR_analysis/ldsc/ldsc_results.txt")
content = file.readlines()

listOfListOfSummResults119Bacs = []
for i in lstOfSumResultsStartIndex:
    listOfListOfSummResults119Bacs.append(content[i:i+3][-1].split())
```

```python
len(listOfListOfSummResults119Bacs)
```

```python
#listOfListOfSummResults119Bacs[0][:2]

for i in range(len(listOfListOfSummResults119Bacs)):
    listOfListOfSummResults119Bacs[i][0] = "%s.summary.sumstats.gz"%(listOfListOfSummResults119Bacs[i][0].split("/")[4].split(".")[3])
    listOfListOfSummResults119Bacs[i][1] = listOfListOfSummResults119Bacs[i][1].split("/")[-1]
```

```python
import pandas as pd

df = pd.DataFrame(listOfListOfSummResults119Bacs, columns = ["p1"	,"p2"	,"rg",	"se"	,"z"	,"p"	,"h2_obs",	"h2_obs_se"	,"h2_int",	"h2_int_se",	"gcov_int"	,"gcov_int_se"])
df
```
