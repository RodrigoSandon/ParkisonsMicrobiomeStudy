```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
def readFilesAndProcess (searchFile, dirToProcess, editedVariantsOutPath, prunedVariantsOutPath):
    d = []
    #load csv of finalMB_Variants
    allMBVariants = pd.read_csv(searchFile)
    
    for root, dirs, files in os.walk(dirToProcess):
        for name in files:
            currPath = os.path.join(root, name)
            if currPath.endswith("prunedVariants.csv"): # now only looking at the variants that were kept
                currPrunedFile = pd.read_csv(currPath) #now have a pruned csv file
                #print(type(currPrunedFile["RS"].tolist()))
                #print(currPrunedFile["RS"].tolist())
                try:
                    lstOfKept = list(currPrunedFile["RS"].tolist()) # make column "RS" as a list
                except KeyError:
                    print("key error at: ", currPath)
                #print(lstOfKept)
                #search algo
                for idxOfRow, series in allMBVariants.iterrows():
                    chromID = series["chrID"]
                    if chromID in lstOfKept:
                        #process -> ad yes to kept?
                        allMBVariants.at[idxOfRow, "Kept?"] = "yes"
                        #print(idxOfRow)
                        #print(allMBVariants.iloc[idxOfRow][allMBVariants.columns[1]])
                        d.append({allMBVariants.columns[0]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[0]], 
                                  allMBVariants.columns[1]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[1]],
                                  allMBVariants.columns[2]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[2]],
                                 allMBVariants.columns[3]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[3]], 
                                  allMBVariants.columns[4]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[4]],
                                  allMBVariants.columns[5]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[5]],
                                 allMBVariants.columns[6]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[6]], 
                                  allMBVariants.columns[7]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[7]],
                                  allMBVariants.columns[8]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[8]],
                                 allMBVariants.columns[9]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[9]], 
                                  allMBVariants.columns[10]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[10]],
                                  allMBVariants.columns[11]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[11]],
                                 allMBVariants.columns[12]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[12]], 
                                  allMBVariants.columns[13]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[13]],
                                  allMBVariants.columns[14]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[14]],
                                 allMBVariants.columns[15]: allMBVariants.iloc[idxOfRow][allMBVariants.columns[15]]})
    #making edited finalMB_Variants file
    allMBVariants.to_csv(editedVariantsOutPath, index = False, header = True)
    #making prunedCSv
    csv_columns = ["rsID",	"chrID",	"Chr",	"start(hg19)",	"stop(hg19)",	"A1",	"A2",	"freq",	"b",	"se",	"p",	"N_cases",	"N_controls",	"Close genes",	"PMID",	"Underlying_studies_PMID"]
    try:
        with open(prunedVariantsOutPath, 'w') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
            writer.writeheader()
            for data in d:
                writer.writerow(data)
    except IOError:
        print("I/O error")
```

```python
def matchAndFill (table1, table2, out):
    df1 = pd.read_csv(table1)
    df2 = pd.read_csv(table2)
    
    #df2.fillna("0") doesn't actually change the df
    for index, row in df1.iterrows():
        snp = row["name"]
        chrom = row["#chrom"].split("r")
        #print(chrom)
        try: 
            chromNum = int(chrom[1])
        except ValueError:
            pass
        for idx, rw in df2.iterrows():
            snpToMatch = rw["rsID"]
            if snp == snpToMatch:
                #print(rw["start(hg19)"])
                df2.at[idx, "Chr"] = chromNum
                df2.at[idx, "start(hg19)"] = row["chromStart"]
                df2.at[idx, "stop(hg19)"] = row["chromEnd"]
                
    df2.to_csv(out, index = False, header = True)
```

```python
table1 = "/Volumes/T7Touch/NIHSummer2021/Data/MicrobiomeVariants/hg19_positions.csv"
table2 = "/Volumes/T7Touch/NIHSummer2021/Data/MicrobiomeVariants/allMB_variants.csv"
out = "/Volumes/T7Touch/NIHSummer2021/Data/MicrobiomeVariants/reformattedMB_variants.csv"

matchAndFill(table1, table2, out)
```

```python
def doubleDict(df, dictName):
    for index, row in df.iterrows():
        chrID = row["SNP"]
        chrom = row["SNP"].split(":")
        #print("here")
        name = chrom[0]
        #print(name)
        dictName[name][chrID] = [row["SNP"], row["A1"], row["A2"], row["freq"], row["b"], row["se"], row["p"], row["N_cases"], row["N_controls"]]
```
